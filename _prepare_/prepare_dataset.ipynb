```python
import os
import numpy as np


datapath = '/data_nas2/pxb/CNX-PathLLM/GTEx-TCGA-Embeddings/'
project_names = ['TCGA-BLCA', 'TCGA-BR', 'TCGA-CESC', 'TCGA-COAD', 
                 'TCGA-GBM', 'TCGA-HNSC', 'TCGA-LGG', 'TCGA-LIHC', 
                 'TCGA-LUNG', 'TCGA-OV', 'TCGA-PRAD', 'TCGA-RCC', 
                 'TCGA-Rest', 'TCGA-SARC', 'TCGA-SKCM', 'TCGA-STAD', 
                 'TCGA-THCA', 'TCGA-UCEC']
for project_name in project_names:
    project_path = os.path.join(datapath, project_name)
    npy_files = os.listdir(project_path)
    
    print(f'Processing {project_name}...')
    num_dict = {'0_1024_d1024': 0,
                '0_1024_d512': 0,
                '1_1024_d1024': 0,
                '1_1024_d512': 0,
                '1_512_d1024': 0,
                '1_512_d512': 0
        }
    for npy_file in npy_files:
        if npy_file.endswith('_0_1024.npy'):
            mid_npy = np.load(os.path.join(project_path, npy_file), allow_pickle=True)
            if mid_npy[()]['feature'].shape[1] == 512:
                num_dict['0_1024_d512'] += 1
            elif mid_npy[()]['feature'].shape[1] == 1024:
                num_dict['0_1024_d1024'] += 1
        elif npy_file.endswith('_1_1024.npy'):
            mid_npy = np.load(os.path.join(project_path, npy_file), allow_pickle=True)
            if mid_npy[()]['feature'].shape[1] == 512:
                num_dict['1_1024_d512'] += 1
            elif mid_npy[()]['feature'].shape[1] == 1024:
                num_dict['1_1024_d1024'] += 1
        elif npy_file.endswith('_1_512.npy'):
            mid_npy = np.load(os.path.join(project_path, npy_file), allow_pickle=True)
            if mid_npy[()]['feature'].shape[1] == 512:
                num_dict['1_512_d512'] += 1
            elif mid_npy[()]['feature'].shape[1] == 1024:
                num_dict['1_512_d1024'] += 1
    print(num_dict)
        
```

```python
import os
import shutil
file_root = '/home/ljs/code/Gene_model/My_PANTHER/PANTHER/src/splits/GTEx-TCGA-class'
file_root_new = '/home/ljs/code/Gene_model/My_PANTHER/PANTHER/src/splits/GTEx-TCGA-copy'


for project_name in os.listdir(file_root):
    print(f'Deal with project {project_name}')
    for mag in os.listdir(os.path.join(file_root, project_name)):
        for pkl_files in os.listdir(os.path.join(file_root, project_name, mag, 'embeddings')):
            if pkl_files.endswith('.pkl'):
                new_dir = os.path.join(file_root_new, project_name, mag, 'embeddings')
                os.makedirs(new_dir, exist_ok=True)
                shutil.copy(os.path.join(file_root, project_name, mag, 'embeddings', pkl_files), new_dir)
        new_dir = os.path.join(file_root_new, project_name, mag)
        shutil.copy(os.path.join(file_root, project_name, mag, 'train.csv'), new_dir)
            
```
